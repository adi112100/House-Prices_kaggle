### One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
colum_to_encode = list(X.select_dtypes(include=['object']).columns)
colum_to_encode.append('MSSubClass')

for i in colum_to_encode:
    if len( df1[i].unique() ) != len( X[i].unique() ):
        print(i)
        print(len( df1[i].unique() ), len( X[i].unique() ))
```

```python
df1['Condition2'] = 'Norm'
X['Condition2'] = 'Norm'
X['Utilities'] = 'AllPub'
X['RoofMatl'] = 'CompShg'
df1['RoofMatl'] = 'CompShg'
X['Heating'] = 'GasA'
df1['Heating'] = 'GasA'
# X[X['HouseStyle']== '2.5Fin']['HouseStyle'] = '1Story' 
for i in range(len(X)):
    if X.loc[i, 'HouseStyle'] == '2.5Fin':
        X.loc[i, 'HouseStyle'] = '1Story'
    if X.loc[i, 'Exterior1st'] == 'Stone' or X.loc[i, 'Exterior1st'] == 'ImStucc':
        X.loc[i, 'Exterior1st'] = 'VinylSd'
    if X.loc[i, 'Exterior2nd'] == 'Other':
        X.loc[i, 'Exterior2nd'] = 'VinylSd'
    if X.loc[i, 'Electrical'] == 'Mix':
        X.loc[i, 'Electrical'] = 'SBrkr'
    if X.loc[i, 'GarageQual'] == 'Ex':
        X.loc[i, 'GarageQual'] = 'TA'
for i in range(len(df1)):
    if df1.loc[i, 'MSSubClass'] == 150:
        df1.loc[i, 'MSSubClass'] = 20
```
